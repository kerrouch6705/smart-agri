## 1. Séparation entre X et y

Dans cette étape, nous allons séparer la dataset en deux parties :

- `X` : les variables explicatives, c'est-à-dire les colonnes utilisées pour faire la prédiction.
- `y` : la variable cible, c'est-à-dire la colonne que le modèle doit prédire.

Dans notre cas :

- `X` contient toutes les colonnes sauf `Irrigation_Need`.
- `y` contient uniquement la colonne `Irrigation_Need`.

```python
X = df.drop("Irrigation_Need", axis=1)
y = df["Irrigation_Need"]

print("Shape de X :", X.shape)
print("Shape de y :", y.shape)
```

## 2. Encodage de la variable cible y

La variable cible `Irrigation_Need` contient des valeurs textuelles :

- `Low`
- `Medium`
- `High`

Il faut donc transformer ces classes en valeurs numériques.

Pour cela, nous utilisons `LabelEncoder`.

```python
encoder = LabelEncoder()
y_encoded = encoder.fit_transform(y)
encoder.classes_
y_encoded[:15]
```

```python
for classe, valeur in zip(encoder.classes_, range(len(encoder.classes_))):
    print(classe, "->", valeur)
```
