```python
df = pd.read_csv('/kaggle/input/datasets/miadul/irrigation-water-requirement-prediction-dataset/irrigation_prediction.csv')
df
```

```python
df = df.drop_duplicates()
df
```

```python
string_cols = df.select_dtypes(include=["object"]).columns

for col in string_cols:
    df[col] = df[col].astype(str).str.strip()

df
```

```python
categorical_cols = df.select_dtypes(include=["object"]).columns

for col in categorical_cols:
    df[col] = df[col].astype(str).str.strip()
```

```python
numeric_conditions = {
    "Soil_pH": (0, 14),
    "Soil_Moisture": (0, 100),
    "Organic_Carbon": (0, 10),
    "Electrical_Conductivity": (0, 10),
    "Temperature_C": (-10, 60),
    "Humidity": (0, 100),
    "Rainfall_mm": (0, 3000),
    "Sunlight_Hours": (0, 24),
    "Wind_Speed_kmh": (0, 150),
    "Field_Area_hectare": (0, 1000),
    "Previous_Irrigation_mm": (0, 500)
}

print("Invalid numerical values check:")

for col, (min_val, max_val) in numeric_conditions.items():
    invalid_rows = df[(df[col] < min_val) | (df[col] > max_val)]
    print(f"{col}: {invalid_rows.shape[0]} invalid rows")
```

```python
df.to_csv("irrigation_prediction_clean.csv", index=False)
```
